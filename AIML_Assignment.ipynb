## 1. Sequence Generation (1 to 10)

```python
print("1 to 10:")
for i in range(1, 11):
    print(i, end=" ")
```

## 2. Even Numbers (1 to 20)

```python
print("\n\nEven numbers from 1 to 20:")
for i in range(1, 21):
    if i % 2 == 0:
        print(i, end=" ")
```

## 3. Dynamic Multiplication Table

```python
n = int(input("\n\nEnter a number for multiplication table: "))

print(f"\nMultiplication Table of {n}:")
for i in range(1, 11):
    print(f"{n} x {i} = {n*i}")
```

## 4. Factorial

```python
n = int(input("\nEnter a number to find factorial: "))

fact = 1
for i in range(1, n+1):
    fact *= i

print("Factorial =", fact)
```

## 5. Fibonacci Sequence

```python
n = int(input("\nEnter number of terms for Fibonacci: "))

a, b = 0, 1
print("Fibonacci Series:")
for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

## 6. Prime Number Validation

```python
n = int(input("\n\nEnter a number to check prime: "))

if n <= 1:
    print("Not Prime")
else:
    for i in range(2, n):
        if n % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime Number")
```

## 7. Integer Reversal

```python
num = int(input("\nEnter a number to reverse: "))

rev = 0
while num > 0:
    digit = num % 10
    rev = rev * 10 + digit

    num //= 10

print("Reversed Number =", rev)
```

## 8. Numerical Palindrome

```python
num = int(input("\nEnter a number to check palindrome: "))
original = num

rev = 0
while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10

if original == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

## 9. Armstrong Number

```python
num = int(input("\nEnter a number to check Armstrong: "))
original = num

sum = 0
while num > 0:
    digit = num % 10
    sum += digit ** 3
    num //= 10

if sum == original:
    print("Armstrong Number")
else:
    print("Not Armstrong")
```

## 10. Euclidean Algorithm (GCD)

```python
a = int(input("\nEnter first number: "))
b = int(input("Enter second number: "))

while b != 0:
    a, b = b, a % b

print("GCD =", a)
```
