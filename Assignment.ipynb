## 1. Circle Geometry Tool (OOP)

```python
import math

class Circle:
    def __init__(self, radius=1.0):
        if radius <= 0:
            raise ValueError("Radius must be positive")
        self.radius = radius

    def compute(self):
        return {
            "area": round(math.pi * self.radius**2, 2),
            "circumference": round(2 * math.pi * self.radius, 2)
        }

print(Circle(5).compute())
```

## 2. Dynamic Profile Builder

```python
class UserProfile:
    def __init__(self, first, last, **kwargs):
        self.profile = {"first": first, "last": last}
        self.profile.update(kwargs)

    def get(self):
        return self.profile

print(UserProfile("Sruthi","Mentor",field="AI").get())
```

## 3. Smart Multiplier

```python
from functools import reduce
def multiply_numbers(*args):
    return reduce(lambda x,y: x*y, args) if args else 0
print(multiply_numbers(2,3,4))
```

## 6. Factorial with Logging

```python
import logging
logging.basicConfig(level=logging.INFO)
def factorial(n):
    logging.info(f"factorial({n})")
    return 1 if n<=1 else n*factorial(n-1)
print(factorial(5))
```

## 7. Statistical Analyzer

```python
import statistics

def analyze(nums):
    return min(nums), max(nums), statistics.mean(nums)

print(analyze([10,20,30,40,50]))
```

## 8. Division API

```python
def divide(a,b):
    try:
        return a/b
    except Exception as e:
        return str(e)

print(divide(10,2))
```

## 9. File Handler

```python
def read_file(file):
    try:
        with open(file,'r') as f:
            return f.read()
    except:
        return "File error"
```

## 10. Custom Exception

```python
class InvalidVoterError(Exception):
    pass

def check(age):
    if age < 18:
        raise InvalidVoterError
    return True

try:
    check(16)
except Exception as e:
    print("Not eligible")
```

## 12. BMI Function

```python
def bmi(w,h):
    """BMI calc"""
    return round(w/(h**2),2)
print(bmi(70,1.75))
```

## 15. ML Scaler

```python
import numpy as np

def scale(data):
    try:
        arr=np.array(data)
        return ((arr-arr.mean())/arr.std()).tolist()
    except:
        return [0]*len(data)

print(scale([10,20,30]))
```
