```python
# 1. Reverse a String
def reverse_string(s):
    return s[::-1]

print('Reverse:', reverse_string('apple'))
```

```python
# 2. Even or Odd
def check_even_odd(num):
    if num % 2 == 0:
        return 'Even'
    else:
        return 'Odd'

print('Number 7 is', check_even_odd(7))
```

```python
# 3. Largest Number (without max)
def largest_number(lst):
    largest = lst[0]
    for num in lst:
        if num > largest:
            largest = num
    return largest

print('Largest:', largest_number([3, 7, 2, 9, 5]))
```

```python
# 4. Sum of List
def sum_list(lst):
    total = 0
    for num in lst:
        total += num
    return total

print('Sum:', sum_list([1, 2, 3, 4, 5]))
```

```python
# 5. Palindrome Check
def is_palindrome(word):
    return word == word[::-1]

print("Is 'radar' palindrome?", is_palindrome('radar'))
```

```python
# 6. Vowel Counter
def count_vowels(sentence):
    vowels = 'aeiouAEIOU'
    count = 0
    for char in sentence:
        if char in vowels:
            count += 1
    return count

print('Vowels:', count_vowels('Hello World'))
```

```python
# 7. FizzBuzz
for i in range(1, 21):
    if i % 3 == 0 and i % 5 == 0:
        print('FizzBuzz')
    elif i % 3 == 0:
        print('Fizz')
    elif i % 5 == 0:
        print('Buzz')
    else:
        print(i)
```

```python
# 8. Factorial
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print('Factorial of 5:', factorial(5))
```

```python
# 9. Remove Duplicates
def remove_duplicates(lst):
    unique_list = []
    for item in lst:
        if item not in unique_list:
            unique_list.append(item)
    return unique_list

print('Unique List:', remove_duplicates([1, 2, 2, 3, 4, 4, 5]))
```

```python
# 10. Multiplication Table
def multiplication_table(n):
    for i in range(1, 11):
        print(f"{n} x {i} = {n*i}")

multiplication_table(5)
```
